```python
#importing neccesary libraries
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
#Name of companies whose stock data we will collect from yahoo finance
companies_dict={
    'Amazon':'AMZN',
    'Apple':'AAPL',
    'Walgreen':'WBA',
    'Northrop Grumman':'NOC',
    'Boeing':'BA',
    'Lockheed Martin':'LMT',
    'McDonalds':'MCD',
    'Intel':'INTC',
    'Navistar':'NAV',
    'IBM':'IBM',
    'Texas Instrument':'TXN',
    'MasterCard':'MA',
    'Microsoft':'MSFT',
    'General Electrics':'GE',
    'American Express':'AXP',
    'Pepsi':'PEP',
    'Coco Cola':'KO',
    'Johnson & Johnson':'JNJ',
    'Toyota':'TM',
    'Honda':'HMC',
    'Mistubishi':'MSBHY',
    'Sony':'SNE',
    'Exxon':'XOM',
    'Chevron':'CVX',
    'Valero Energy':'VLO',
    'Ford':'F',
    'Bank of America':'BAC'
}
#sorting companies on the basis of their label
companies=sorted(companies_dict.items(),key=lambda x:x[1] )
print(companies)
```

```python
#describing datasource
data_source='yahoo'
#describing duration for which we require data
start_date='2016-01-01'
end_date='2019-12-31'

#extracting the data from yahoo finance
panel_data=data.DataReader(list(companies_dict.values()),data_source,start_date,end_date)
print(panel_data)
```

```python
#converting stock sopen and close to numpy array as it will be easy to perform operation
stock_open=panel_data['Open']
stock_close=panel_data['Close']

#visualizing array
print(stock_close.iloc[0])
```

```python
stock_close=np.array(stock_close).T
stock_open=np.array(stock_open).T

row,col=stock_close.shape
print(row)
print(col)
```

```python
#initializing movements for stroing change in stock market for rach company each day
movements=np.zeros([row,col])

#calcumg movements
for i in range(0,row):
    movements[i,:]=np.subtract(stock_close[i,:],stock_open[i,:])
    
#visualizing data    
for i in range(0,len(companies)):
    print('Company {} ; Change {}'.format(companies[i][0],sum(movements[i][:])))
    
#By seeing the data it can be said that whose sum of change is positive those are safe to buy    
```

```python
#conparing the movements of amazon and apple
plt.clf
plt.figure(figsize=(18,16))
axl=plt.subplot(221)
plt.plot(movements[1][:])
plt.title(companies[1])

plt.subplot(222,sharey=axl)
plt.plot(movements[0][:])
plt.title(companies[0])
plt.show()

#this plot the great variation in movement in order to make things easy we have to normalize it to compare it on same scale 
```

```python
#importing normalizer
from sklearn.preprocessing import Normalizer
#initializing normalizer
normalizer=Normalizer()
#fiting and tranforming data to normalize
new=normalizer.fit_transform(movements)

#Analyzing the normalize data
print(new.max())
print(new.min())
print(new.mean())
```

```python
#Now analyzing the movements after normalizing
plt.clf
plt.figure(figsize=(18,16))
axl=plt.subplot(221)
plt.plot(new[1][:])
plt.title(companies[1])

plt.subplot(222,sharey=axl)
plt.plot(new[0][:])
plt.title(companies[0])
plt.show()
#this plot shows this data can be compare easily as range has become short
```

```python
#importing pipeline and kmeans clustering
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans
```

```python
#initializing kmeans and normalizer and fit data 
kmeans=KMeans(n_clusters=10,max_iter=1000)
pipeline=make_pipeline(normalizer,kmeans)
pipeline.fit(movements)
#this can show the pipeline created
```

```python
#View metrics of kmeans
print(kmeans.inertia_)
```

```python
#predicting the different clusters
labels=pipeline.predict(movements)
df=pd.DataFrame({'Labels':labels,'Companies':companies})

print(df.sort_values('Labels'))
```

```python
# importing and implementing PCA for dimensionality reduction
from sklearn.decomposition import PCA

reduced_data=PCA(n_components=2).fit_transform(new)

kmeans=KMeans(n_clusters=10)
kmeans.fit(reduced_data)
```

```python

#predicting the different clusters after dimensionality reduction
labels=kmeans.predict(reduced_data)
df=pd.DataFrame({'Labels':labels,'Companies':companies})

print(df.sort_values('Labels'))
```

```python
h = 0.01
# Plot the decision boundary
x_min,x_max = reduced_data[:,0].min()-1, reduced_data[:,0].max() + 1
y_min,y_max = reduced_data[:,1].min()-1, reduced_data[:,1].max() + 1
xx,yy = np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
# Obtain labels for each point in the mesh using our trained model
Z = kmeans.predict(np.c_[xx.ravel(),yy.ravel()])
# Put the result into a color plot
Z = Z.reshape(xx.shape)
# Define color plot
cmap = plt.cm.Paired
# Plotting figure
plt.clf()
plt.figure(figsize=(10,10))
plt.imshow(Z,interpolation ='nearest',extent=(xx.min(),xx.max(),yy.min(),yy.max()),cmap = cmap,aspect = 'auto',origin = 'lower')
plt.plot(reduced_data[:,0],reduced_data[:,1],
         'k.',markersize = 5)
# Plot the centroid of each cluster as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0],centroids[:,1],marker = 'x',s = 169,linewidths = 3,color = 'w',zorder = 10)
plt.title('K-Means clustering on stock market movements (PCA-Reduced data)')
plt.xlim(x_min,x_max)
plt.ylim(y_min,y_max)
plt.show()
```
